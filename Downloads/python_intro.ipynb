## Основы `Python` и `Jupyter`

### Полезные сведения о `Jupyter`

Ноутбук `Jupyter` это интерактивная среда для языка `Python`.\
Ноутбук состоит из ячеек двух типов: содержащих исполняемый код или текст с поддержкой `Markdown` и $\LaTeX$.\
Чтобы переключить ячейку в текстовый режим нужно выделить её и нажать `M`.\
Чтобы переключить ячейку в режим кода нужно выделить её и нажать `Y`.\
Чтобы выполнить код в ячейке или отформатировать текстовую ячейку нужно нажать `Shift+Enter`.\
Результаты выполнения кода, такие как объявленные переменные и функции, сохраняются между вызовами ячеек в глобальной области видимости и доступны всем ячейкам до остановки сервера или ядра.\
После выполнения ячейки результаты могут выпечатываться непосредственно под ней.\
Двойной клик по ячейке или нажатие `Enter` для выделенной ячейки переведёт её в режим редактирования.\
И, наоборот, нажатие `Esc` в режиме редактирование переведёт ячейку в режим выделения.\
Для навигации по ячейкам в режиме выделения можно использовать стрелки.\
Можно выделять несколько ячеек, зажав `Shift` и используя стрелки или нажатия мыши.\
В правом верхнем углу каждой ячейки есть кнопки позволяющие создавать новые ячейки и перемещать текущую.\
Создавать ячейки в режиме выделения можно также по нажатию `A` или `B`.\
Удалять ячейки в режиме выделения можно нажав `X`.\
Переименовать ноутбук можно через меню `File -> Rename` или в левой панели.\
Не забывайте сохранять ноутбук перед закрытием страницы, и, более важно, перед остановкой сервера в терминале.

<span style="color:red; font-weight: bold">ВАЖНО!!</span>

Все объекты в `Python` можно разделить на две группы: **изменяемые (mutable)** и **неизменяемые (immutable)**.\
На один и тот изменяемый объект может указывать несколько переменных!\
Так происходит потому что переменные это всего лишь ссылки на объекты.

```python
print('Variable l stores a list:')
print(l)
print()

k = l

print('Variable k now stores same list:')
print(k)
print()

k[-1] = 'changed!'

print('We changed list using variable k:')
print(k)
print()

print('But list in variable l is also changed:')
print(l)
print()
```

Для создания `shallow`-копии списка используйте слайс.\
В данном примере переменная `k` указывает уже на другой список.

```python
k = l[:]

print('l =',l)
print('k =',k)

k[-1] = 'Changed again!'

print('l =',l)
print('k =',k)
```

Однако, если список содержит другие списки, то `shallow`-копии может быть недостаточно. Позднее будет показано как создать полную копию.\
Любой объект за исключением логических, числовых типов, значения `None`, строк и кортежей (`tuple`) является **изменяемым**!\
Если вы не уверены является ли какой то объект изменяемым вы можете быстро это проверить самостоятельно в коде.

```python
k.append([1,2,3,4,5])
n = k[:]
print('k =',k)
print('n =',n)

print('\nChanged value of regular element in shallow copy...\n')
n[2] = 'changed!'

print('k =',k)
print('n =',n)

print('\nChanged value of inner list in shallow copy...\n')
n[-1][3] = 'changed!'

print('k =',k)
print('n =',n)
```

Также не всегда очевидно будет ли создана копия при выполнении той или иной операции.\
В таких случаях лучше написать тестовый код перед использованием.\
Смотрите пример ниже.

```python
l = [1,2,3,4,5]
k = [[10,11],[12,13]]

# assign by slice
l[1:2] = k
print('l =',l)

# change list value in 'k' variable
print('\nChanged k...\n')
k[0][1] = 7

# and see that l is also changed!
print('l =',l)
```

#### Функции

Часто повторяющийся код можно записать в виде функции и вызывать его в различных частях программы.

```python
# let's suppose that function receives a string with '0' and '1'
def count_bits(value):
        
    # you can declare local variables which are not visible outside of function
    # in general it is better to use local variables than globally defined
    total_bits = 0
    
    for char in value:
        if char == '1':
            total_bits += 1
    
    # this value will be returned when function is called
    return total_bits
```

Пример: вызов функции с разным аргументом дважды, формирующий кортеж из двух значений.

```python
count_bits('100110101101'), count_bits('110100')
```

В этом примере функция принимает больше аргументов.\
Опционально можно указать ожидаемые типы аргументов и тип возвращаемого значения.\
Также можно указать значения по умолчанию для аргументов.

```python
def make_basis_spin_half(sites: int, spinsup: int = None) -> list:
    
    basis = []
    full_basis_size = 1<<sites
    
    template = '{{0:0{0}b}}'.format(sites)
    if spinsup is None:
        for i in range(full_basis_size):
            el = template.format(i)
            basis.append([int(b) for b in el])
    else:
        for i in range(full_basis_size):
            el = template.format(i)
            if count_bits(el) == spinsup:  # here we use function defined above
                basis.append([int(b) for b in el])
    
    return basis
```

Вызов функции с аргументом по умолчанию для построения полного базиса.

```python
full_basis = make_basis_spin_half(4)
full_basis
```

Указание полной намагниченности для построения базиса в секторе с заданным полным спином.

```python
basis = make_basis_spin_half(4, 2)
basis
```

Используя метод `join()` строкового объекта и конструктор списка можно конвертировать списки в строки и затем объединить их снова для получения списка значений разделённых запятыми (`CSV`).

```python
states = []
for state in basis:
    states.append(''.join(str(b) for b in state))
print(', '.join(states))
```

#### Ввод / вывод

В `Python` есть методы для записи данных в файлы чтобы сохранить результаты расчётов между запусками ноутбука или для их перемещения на другой компьютер.

Открытие простого текстового файла.\
Инструкция `open()` принимает два аргумента: имя файла и режим доступа.\
В примере ниже создаётся файл с именем `basis.dat` в режиме для записи `'w'` который создаёт новый или перезаписывает существующий файл.

```python
f = open('basis.dat', 'w')
```

Далее мы записываем наш базис в одну колонку

```python
for state in states:
    f.write('{0}\n'.format(state))
```

При закрытии файла система принудительно записывает данные в файл на диск.\
До момента закрытия данные могу находиться в оперативной памяти.

```python
f.close()
```

Теперь можно открыть проводник или другой файловый менеджер и увидеть там созданный файл. 

Для удобства работы с файлами можно использовать менеджер контекста.\
В этом случае файл `f` будет автоматически закрыт при выходе из соответствующего блока или при возниконвении ошибки.

```python
with open('full_basis.dat', 'w') as f:
    for state in full_basis:
        f.write('{0}\n'.format(''.join(str(b) for b in state)))
```

#### Основы ООП в `Python`

Помимо базовых типов и объектов `Python` разработчик может определять свои типы. В разделе выше мы использовали объект `np.ndarray` который не является базовым типом языка. Здесь мы покажем как создавать собственные типы. Для этого нужно определить "чертёж", по которому будут создаваться объекты данного типа. Такой "чертёж" называется классом и определяется через ключевое слово `class`. Ниже приводится пример создания класса для точечной частицы, двигающейся в гравитационном поле Земли. Все инструкции будут пояснены ниже.  

```python
from scipy.constants import g
```

```python
class PointParticle:
    
    def __init__(self, coordinates: np.ndarray, velocity: np.ndarray, mass: float):
        
        self.coordinates = coordinates
        self.velocity = velocity
        self.mass = mass
        
    def __str__(self):
        return f'PointParticle<x={self.coordinates},v={self.velocity},m={self.mass}>'
    
    def __repr__(self):
        return self.__str__()
    
    def kinetic_energy(self):
        return self.mass*np.dot(self.velocity, self.velocity)/2
    
    def potential_energy(self):
        return self.mass*g*self.coordinates[2]
    
    def full_energy(self):
        return self.kinetic_energy() + self.potential_energy()
    
    def distance(self,particle):
        return np.linalg.norm(self.coordinates - particle.coordinates)
```

Здесь мы сделали следующее:

1. Определили имя нашего класса как `PointParticle` при помощи кторого мы сможем создавать объекты (инстансы) данного класса вызвав его конструктор `PointParticle(coordinates, velocity, acceleration, mass)`.
2. Все функции, определённые внутри блока `class` будут связаны с конкретными объектами данного класса и называются методами. Вы можете заметить, что методы принимают `self` в качестве первого аргумента. Эта переменная ссылается на объект класса. Используя эту переменную можно изменять состояние объекта.
3. Первый метод который мы определели в классе это конструктор `__init__()` который определяет как создаётся объект. Здесь мы используем переменную `self` чтобы добавлять данные к создаваемому объекту задавая координаты, скорость и т.д. Этот метод вызывается при создании объекта `PointParticle(coordinates, velocity, acceleration, mass)` и возвращает объект данного класса. Заметьте, что здесь отсутствует инструкция `return` - конструктор возвращает значение автоматически. Мы также указали ожидаемые типы аргументов в функциях. Они используютя средой `Jupyter` и другими `IDE` для автодополнения и подсказок.
4. Следующие два метода `__str__()` и `__repr__()` определяют как наш объект быдет преобразовываться в строку при выпечатывании в консоль или при выводе в файл. Здесь можно заметить, что переменная `self` может использоваться для вызова методов нашего класса.
5. Следующие три метода рассчитывают кинетическую, потенциальную и полную энергии частицы.
6. Наконец, итоговый метод позволяет расчитать расстояние между частицами.

Создадим объект нашего класса.

```python
particle1 = PointParticle(
    np.array([0,1,1]),
    np.array([0,1/np.sqrt(2),1/np.sqrt(2)]),
    1.0
)
particle1
```

Теперь мы можем вызывать методы для конкретного объекта:

```python
particle1.full_energy()
```

```python
particle1.kinetic_energy() + particle1.potential_energy()
```

Важно, что мы можем создавать несколько экземпляров класса и они не будут зависеть друг от друга:

```python
particle2 = PointParticle(
    np.array([1,1,0]),
    np.array([0,2,3]),
    3.0
)
particle2
```

```python
particle2.full_energy()
```

Ограничений на использование нашего объекта почти не существует. Например, мы можем создать массив частиц, представляющий одномерную цепочку.

```python
m = 1
particles = [PointParticle(np.array([i,0,0]), np.array([0,0,0]), m) for i in range(10)]
particles
```

Новые классы могут быть определены при помощи механизма наследования. Предположим, мы хотим создать класс, представляющий сферические частицы конечного объёма. В то же время наш коллега может захотеть определить класс заряженных точечных частиц. Оба типа частиц обладают одинаковым набором характеристик (позиция, скорость, масса) и добавляют новые. Без наследования пришлось бы повторять один и тот же код для обоих классов. Но вместо этого мы можем использовать наследование следующим образом:

```python
class SphericalParticle(PointParticle):
    
    def __init__(self, coordinates: np.ndarray, velocity: np.ndarray, mass: float, radius: float):
        
        super(SphericalParticle, self).__init__(coordinates, velocity, mass)
        
        self.radius = radius
        
    def __str__(self):
        return f'SphericalParticle<x={self.coordinates},v={self.velocity},m={self.mass},r={self.radius}>'
    
    def volume(self):
        return 4/3*np.pi*self.radius**3
```

```python
class ChargedParticle(PointParticle):
    
    def __init__(self, coordinates: np.ndarray, velocity: np.ndarray, mass: float, q: float):
        
        super(ChargedParticle, self).__init__(coordinates, velocity, mass)
        
        self.q = q
        
    def __str__(self):
        return f'ChargedParticle<x={self.coordinates},v={self.velocity},m={self.mass},q={self.q}>'
    
    def full_energy(self, particles: list = []):
        result = super(ChargedParticle, self).full_energy()
        for p in particles:
            result += self.q * p.q / self.distance(p)
        return result
```

Здесь мы сделали следующее:

1. При определении новых классов мы указали базовый класс `PointParticle`. Это приведёт к наследованию всех свойств и методов класса точечной частицы для новых классов.
2. Используя функцию `super()` мы получили ссылку на базовый класс и вызвали его конструктор. Это действие инициализирует поля базового класса.
3. Переопределили метод `__str()__` для каждого класса.
4. Для класса `SphericalParticle` добавили метод, вычисляющий объём.
5. Для класса `ChargedParticle` переопределили метод вычисляющий полную энергию. Заметьте, как мы вновь используем метод `super()` чтобы получить объект базового класса так как название метода `full_energy()` одинаково в базовом и производном классе.

Создадим объект сферической частицы:

```python
sp = SphericalParticle(
    np.array([0,1,1]),
    np.array([0,1/np.sqrt(2),1/np.sqrt(2)]),
    7.0,
    1.0
)
sp
```

Теперь мы можем использовать новый метод для расчёта её объёма.

```python
sp.volume()
```

И мы можем использовать и методы базового класса для расчёта энергии:

```python
sp.full_energy()
```

```python
particles_num = 10

charged_particles = []

coordinates = 2*gen.random((particles_num,3)) - 1
coordinates[:,2] = np.abs(coordinates[:,2])
velocities = 2*gen.random((particles_num,3)) - 1
masses = 3*gen.random((particles_num,))
charges = 2*gen.random((particles_num,)) - 1

for i, (x,v,m,q) in enumerate(zip(coordinates,velocities,masses,charges)):
    charged_particles.append(ChargedParticle(x,v,m,q))
charged_particles
```
